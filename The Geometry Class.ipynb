We have a .txt file containing lines of instructions.

Each lines contains a instruction(1~5), each instruction has it's own purpose.

1: Initialize a specific shape according to given shape class, its id, and the edge length parameter.

2: Set its edge/radius according to id and given edge/radius length.

3: Print its perimeter by its id.

4: Print its area by its id.

5: Sort all shapes in the list by area and print their ids out.

* There are only 3 specific shapes: Circle, Square and Regular Triangle
* Instruction 1: 1,"Shape Class","id","edge/radius length"
* Instruction 2: 2,"id","edge/radius length"
* Instruction 3: 3,"id"
* Instruction 4: 4,"id"
* Instruction 5: 5

```python
#!/usr/bin/python3

"""
inputdata3.txt:
1,Circle,0,5
1,RegularTriangle,1,3
1,Square,2,10
2,2,1
2,0,1
3,0
4,2
5
"""

from math import pi, sqrt


class Shape(object):

    def __init__(self,  identification_number, edge):
        self.edge = float(edge)
        self.id =  identification_number

    def get_area(self):
        pass

    def get_perimeter(self):
        pass

    def set_edge(self, edge):
        self.edge = float(edge)

    def __lt__(self, other):
        return self.get_area() < other.get_area()

    def __str__(self):
        return self.id

class Circle(Shape):
    def __init__(self,  identification_number, edge):
        super().__init__( identification_number, edge)

    def get_area(self):
        area = round(pi * self.edge**2, 2)
        return area

    def get_perimeter(self):
        perimeter = round(pi * self.edge * 2, 2)
        return perimeter

class Square(Shape):
    def __init__(self,  identification_number, edge):
        super().__init__( identification_number, edge)

    def get_area(self):
        area = round(self.edge**2, 2)
        return area

    def get_perimeter(self):
        perimeter = round(self.edge * 4, 2)
        return perimeter

class RegularTri(Shape):
    def __init__(self,  identification_number, edge):
        super().__init__( identification_number, edge)

    def get_area(self):
        area = round(sqrt(3) / 4 * self.edge**2, 2)
        return area

    def get_perimeter(self):
        perimeter = round(self.edge * 3, 2)
        return perimeter


if __name__ == "__main__":
    f = open("inputdata3.txt", "r")
    shape_list = []
    for line in f.readlines():
        target = line.strip().split(",")
        if target[0] == "1":
            if target[1] == "Circle":
                obj = Circle(target[2], target[3])
            elif target[1] == "RegularTriangle":
                obj = RegularTri(target[2], target[3])
            elif target[1] == "Square":
                obj = Square(target[2], target[3])
            shape_list.append(obj)
        elif target[0] == "2":
            for obj in shape_list:
                if obj.id == target[1]:
                    obj.set_edge(target[2])
        elif target[0] == "3":
            for obj in shape_list:
                if obj.id == target[1]:
                    print("%.2f" % obj.get_perimeter())
        elif target[0] == "4":
            for obj in shape_list:
                if obj.id == target[1]:
                    print("%.2f" % obj.get_area())
        elif target[0] == "5":
            shape_list.sort()
            for obj in shape_list:
                print(obj)
```
